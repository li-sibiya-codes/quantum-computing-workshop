```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit_aer import AerSimulator
import matplotlib.pyplot as mpl
from qiskit.circuit.library import MCXGate
```

```python
qB = QuantumRegister(4)
cB = ClassicalRegister(3)
qcB = QuantumCircuit(qB,cB)

qcB.h(qB[0])
qcB.h(qB[1])
qcB.h(qB[2])
qcB.x(qB[3])
qcB.h(qB[3])
qcB.barrier()
qcB.x(qB[0])
qcB.x(qB[1])
qcB.x(qB[2])
qcB.mcx([0,1,2], 3)
qcB.x(qB[0])
qcB.x(qB[1])
qcB.x(qB[2])
qcB.cx(qB[2], qB[3])
qcB.cx(qB[1], qB[3])
qcB.ccx(qB[1],qB[2], qB[3])
qcB.cx(qB[0], qB[3])
qcB.ccx(qB[0], qB[2], qB[3])
qcB.ccx(qB[0], qB[1], qB[3])
qcB.mcx([0,1,2], 3)
qcB.barrier()
qcB.h(qB[0])
qcB.h(qB[1])
qcB.h(qB[2])
qcB.measure(qB[0],cB[0])
qcB.measure(qB[1],cB[1])
qcB.measure(qB[2],cB[2])

qcB.draw('mpl')
```

```python
simulator = AerSimulator()
result=simulator.run(qcB).result()
counts = result.get_counts()
print(counts)
```

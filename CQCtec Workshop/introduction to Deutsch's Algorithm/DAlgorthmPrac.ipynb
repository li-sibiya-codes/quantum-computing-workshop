```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit_aer import AerSimulator
import matplotlib.pyplot as mpl
```

```python
#Implementation of Deutsch's algorithm
#constant function 0: input = 0, f(0) = 0

q = QuantumRegister(2)
c = ClassicalRegister(1)
qc = QuantumCircuit(q,c)

qc.h(q[0])
qc.x(q[1])
qc.h(q[1])
qc.barrier()
qc.barrier()
qc.h(q[0])
qc.measure(q[0],c[0])

qc.draw('mpl')
```

```python
simulator = AerSimulator()
result=simulator.run(qc).result()
counts = result.get_counts()
print(counts)
```

```python
#Implementation of Deutsch's algorithm
#constant function 1: input = 1, f(1) = 1

qa = QuantumRegister(2)
ca = ClassicalRegister(1)
qca = QuantumCircuit(qa,ca)

qca.x(qa[0])
qca.h(qa[0])
qca.x(qa[1])
qca.h(qa[1])
qca.barrier()
qca.x(qa[1])
qca.barrier()
qca.h(qa[0])
qca.measure(qa[0],ca[0])

qca.draw('mpl')
```

```python
simulator = AerSimulator()
result=simulator.run(qca).result()
counts = result.get_counts()
print(counts)
```

```python
#Implementation of Deutsch's algorithm
#balanced function x: input = 0 or 1, f(0) = 1 and f(1)=0

qb = QuantumRegister(2)
cb = ClassicalRegister(1)
qcb = QuantumCircuit(qb,cb)


qcb.h(qb[0])
qcb.x(qb[1])
qcb.h(qb[1])
qcb.barrier()
qcb.x(qb[0])
qcb.cx(qb[0],qb[1])
qcb.x(qb[0])
qcb.barrier()
qcb.h(qb[0])
qcb.measure(qb[0],cb[0])

qcb.draw('mpl')
```

```python
simulator = AerSimulator()
result=simulator.run(qcb).result()
counts = result.get_counts()
print(counts)
```
